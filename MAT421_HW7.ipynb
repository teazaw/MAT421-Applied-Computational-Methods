# **Numerical Differentiation Problem Statement**

The **numerical differentation problem statement** refers to the challenge of approximating the derivative of a function at a given point using only a set of discrete function values, typically obtained by sampling the function at specific points. Since we often only function values at discrete points, we need approximately methods to estimate derivatives.

These approximation methods include:
1. Forward difference
2. Backward difference
3. Central difference

The following example shows how numerical differentiation can approximte derivatives when only discrete function values are available. The central difference method is generally the most accurate.

```python
import numpy as np
import matplotlib.pyplot as plt

# define the function
def f(x):
    return np.sin(x)

def f_derivative(x):
    return np.cos(x)

# generate a numerical grid
a, b = 0, 2*np.pi
n = 100
x = np.linspace(a, b, n)
h = x[1] - x[0]

# numerical differentiation methods
forward_diff = (f(x[1:]) - f(x[:-1])) / h
backward_diff = (f(x[1:]) - f(x[:-1])) / h
central_diff = (f(x[2:]) - f(x[:-2])) / (2*h)

#exact derivative
exact_derivative = f_derivative(x)


plt.figure(figsize=(10, 5))
plt.plot(x, exact_derivative, label="Analytical Derivative", linestyle="dashed", color="black")
plt.plot(x[:-1], forward_diff, label="Forward Difference", linestyle="dotted", color="blue")
plt.plot(x[1:], backward_diff, label="Backward Difference", linestyle="dotted", color="red")
plt.plot(x[1:-1], central_diff, label="Central Difference", linestyle="solid", color="green")
plt.xlabel("x")
plt.ylabel("Derivative Approximation")
plt.legend()
plt.title("Numerical Differentiation of sin(x)")
plt.show()
```

# **Finite Difference Approximating Derivatives**

**Finite difference methods** are numerical techniques used to approximate derivatives of functions using discrete points. These approximations are crucial when the function is only known at specific values, such as experimental or computational data.

The **forward difference** is to estimate the slope of the function at *xj* using the line that connects (*xj, f(xj)*) and (*xj+1, f(xj+1)*)

The **backward difference** is to estimate the slope of the function at *xj* using the line that connects (*xj, f(xj)*) and (*xj-1, f(xj-1)*)

The **central difference** is to estimate the slope of the function at *xj* using the line that connects (*xj-1, f(xj-1)*) and (*xj+1, f(xj+1)*)

The following figure demonstrates the different type of formulas to estimate the slope.

image.png


The following example uses the differentiation methods to estimate the slope of a function.

```python
import numpy as np
import matplotlib.pyplot as plt

#define function
def f(x):
    return np.exp(x)

def f_derivative(x):
    return np.exp(x)


a, b = 0, 2
n = 100
x = np.linspace(a, b, n)
h = x[1] - x[0]


forward_diff = (f(x[1:]) - f(x[:-1])) / h  # Forward difference
backward_diff = (f(x[1:]) - f(x[:-1])) / h  # Backward difference
central_diff = (f(x[2:]) - f(x[:-2])) / (2*h)  # Central difference


exact_derivative = f_derivative(x)


plt.figure(figsize=(10, 5))
plt.plot(x, exact_derivative, label="Analytical Derivative", linestyle="dashed", color="black")
plt.plot(x[:-1], forward_diff, label="Forward Difference", linestyle="dotted", color="blue")
plt.plot(x[1:], backward_diff, label="Backward Difference", linestyle="dotted", color="red")
plt.plot(x[1:-1], central_diff, label="Central Difference", linestyle="solid", color="green")

plt.xlabel("x")
plt.ylabel("Derivative Approximation")
plt.legend()
plt.title("Finite Difference Approximations of e^x")
plt.show()

```

# **Approximating of Higher Order Derivatives**

**Higher-order derivatives** refer to the derivatives of a function taken multiple times. The **first derivative** describes the rate of change of a function, and the **second derivative** describes the curvature of the function.

A **Taylor series** is a way of approximating a function as an infinite sum of terms based on its derivatives at a single point. For example, taking the Taylor series around *a = xj* gives

The following example uses a Taylor series to approximate sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt


def taylor_series_sin(x, n_terms=5):
    approximation = 0
    for n in range(n_terms):
        term = ((-1)**n) * (x**(2*n+1)) / np.math.factorial(2*n+1)
        approximation += term
    return approximation


x = np.linspace(-2 * np.pi, 2 * np.pi, 400)


y_exact = np.sin(x)

# taylor series approximation
y_taylor_1 = taylor_series_sin(x, n_terms=1)
y_taylor_3 = taylor_series_sin(x, n_terms=3)
y_taylor_5 = taylor_series_sin(x, n_terms=5)
y_taylor_7 = taylor_series_sin(x, n_terms=7)


plt.figure(figsize=(10, 6))
plt.plot(x, y_exact, label="Exact sin(x)", color="black", linestyle="--")
plt.plot(x, y_taylor_1, label="1 Term Taylor Series", linestyle="--", color="blue")
plt.plot(x, y_taylor_3, label="3 Terms Taylor Series", linestyle=":", color="green")
plt.plot(x, y_taylor_5, label="5 Terms Taylor Series", linestyle="-", color="red")
plt.plot(x, y_taylor_7, label="7 Terms Taylor Series", linestyle="-.", color="purple")

plt.title("Taylor Series Approximation of sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()


```
